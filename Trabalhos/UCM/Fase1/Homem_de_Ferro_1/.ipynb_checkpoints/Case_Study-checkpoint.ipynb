```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
jupyterhub = '/dados/home/joaoviniciuspr'
jupyterlab = '/Users/jjoao/OneDrive/Documentos/Faculdade'
ju = jupyterhub
directory = ju+'/IC_Redes/Trabalhos/UCM/Fase1/Homem_de_Ferro_1'
```

```python
degree=pd.read_csv(directory+'/Degree.csv')
degree=degree.set_index("ID")
degree.fillna('',inplace=True)
count = degree.groupby("Degree").size()
```

```python
fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].plot(count.index,count.values,"o")
ax[0].set_xlabel("Personagem")
ax[0].set_ylabel("Grau")
ax[0].set_title("Gr√°fico de Grau por Personagem")
```

```python
degree=pd.read_csv(directory+'/Degree.csv',dtype=str)
degree=degree.set_index("ID")
degree.fillna('',inplace=True)
degree = degree["Degree"]
for i in degree.index:
    degree[i] = int(degree[i])
degree = degree.sort_values(ascending=False)
fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].plot(degree.values,"o")
ax[0].set_xlabel("Personagens")
ax[0].set_ylabel("Grau")
ax[0].set_title("Gr√°fico de Grau por Personagem")
ax[1].plot(np.log(list(degree.values)),"o")
ax[1].set_xlabel("Personagens")
ax[1].set_ylabel("log(Grau)")
ax[1].set_title("Gr√°fico de log(Grau) por Personagem")
plt.plot()
```
