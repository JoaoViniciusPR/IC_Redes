```python
import pandas as pd
```

# Classe - Tabela de Relações

```python
class Relacao:
    def __init__(self,nomeCena):
        self.Cena=pd.read_excel(nomeCena+' Cena.xlsx')
        self.Cena.fillna('',inplace=True) #Printar mensagem de erro
        d={'Peso':['']}
        self.DF=pd.DataFrame(d,columns=['Peso'],index=[0])
        d1={'Origem':[''],'Destino':[''],'Peso':['']}
        self.relacao=pd.DataFrame(d1,columns=['Origem','Destino','Peso'],index=['0'])
        self.peso_atual=0
        self.executado=0
        
    def data(self):
        return self.relacao
    
    def execut(self):
        if self.executado==1:
            return 'Já executado'
        for i in range(self.Cena.shape[0]-2):
            self.peso_atual=((self.Cena.iloc[i,2]-self.Cena.iloc[i,0])*60+(self.Cena.iloc[i,3]-self.Cena.iloc[i,1])+1)/60
            for j in range(4,self.Cena.shape[1]-1):
                if self.Cena.iloc[i,j]=='':
                    continue
                k=j+1
                while self.Cena.iloc[i,k]!='':
                    rel=str(int(self.Cena.iloc[i,j]))+' '+str(int(self.Cena.iloc[i,k]))
                    if self.DF.shape[0]==1:
                        self.DF.loc[rel,'Peso']=self.peso_atual
                    else:
                        v=sum(self.DF.index==rel)
                        if v==0:
                            self.DF.loc[rel,'Peso']=self.peso_atual
                        else:
                            self.DF.loc[rel,'Peso']+=self.peso_atual
                    k+=1
                    if k>=self.DF.shape[1]:
                        break
        for l in range(1,self.DF.shape[0]-1):
            ind=self.DF.index[l].split()
            self.relacao.loc[l]=[ind[0],ind[1],self.DF.iloc[l,0]]
        self.executado=1
        
    def excel(self,nome):
        self.relacao.to_excel(nome+'_Gephi_Aresta.xlsx')
```

```python
IJ=Relacao('Capitao America 1')
```
